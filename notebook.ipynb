### Cleaning Data

```python
#menangani missing value PM2_5
#menampilkan data yang mengalami missing value
aotizhongxin_df[aotizhongxin_df.PM2_5.isna()]
```

```python
#mencari nilai mean dari pm2_5
aotizhongxin_mean = aotizhongxin_df.PM2_5.mean()
```

```python
#mengganti missing value dengan mean dikarenakan data kontinyu
aotizhongxin_df.PM2_5.fillna(value= aotizhongxin_mean, inplace=True)
```

```python
#memeriksa
aotizhongxin_df.isna().sum()
```

```python
#menangani missing value PM10
#menampilkan data yang mengalami missing value
aotizhongxin_df[aotizhongxin_df.PM10.isna()]
```

```python
#mencari nilai mean dari pm2_5
aotizhongxin_PM10_mean = aotizhongxin_df.PM10.mean()
```

```python
#mengganti missing value dengan mean dikarenakan data kontinyu
aotizhongxin_df.PM10.fillna(value= aotizhongxin_PM10_mean, inplace=True)
```

```python
#memeriksa
aotizhongxin_df.isna().sum()
```

```python
#mengganti missing value yang lain
aotizhongxin_df.SO2.fillna(value= aotizhongxin_df.SO2.mean(), inplace=True)
aotizhongxin_df.NO2.fillna(value= aotizhongxin_df.NO2.mean(), inplace=True)
aotizhongxin_df.CO.fillna(value= aotizhongxin_df.CO.mean(), inplace=True)
aotizhongxin_df.O3.fillna(value= aotizhongxin_df.O3.mean(), inplace=True)
aotizhongxin_df.TEMP.fillna(value= aotizhongxin_df.TEMP.mean(), inplace=True)
aotizhongxin_df.PRES.fillna(value= aotizhongxin_df.PRES.mean(), inplace=True)
aotizhongxin_df.DEWP.fillna(value= aotizhongxin_df.DEWP.mean(), inplace=True)
aotizhongxin_df.RAIN.fillna(value= aotizhongxin_df.RAIN.mean(), inplace=True)
aotizhongxin_df.WSPM.fillna(value= aotizhongxin_df.WSPM.mean(), inplace=True)


```

```python
#memeriksa
aotizhongxin_df.isna().sum()
```

```python
#mengganti kolom wd dengan nilai baris sebelumnya
aotizhongxin_df.wd.fillna(method='ffill', inplace=True)

```

```python
#memeriksa
aotizhongxin_df.isna().sum()
```
