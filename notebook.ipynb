## Exploratory Data Analysis (EDA)

### Explore ...

```python
#menentukan nilai Air Quality Index
#mengubah berdasarkan pedoman dari EPA
#membuat tuple yang berisi : 2 nilai paling depan berisi nilai dari polutan
#2 nilai paling belakang berisi index AQI
PM2_5_breakpoints = [(0, 12, 0, 50), (13, 35, 51, 100), (36, 55, 101, 150), 
                     (56, 150, 151, 200), (151, 250, 201, 300), (251, 9999, 301, 500)]
PM10_breakpoints = [(0, 54, 0, 50), (55, 154, 51, 100), (155, 254, 101, 150), 
                     (255, 354, 151, 200), (355, 424, 201, 300), (425, 9999, 301, 500)]
SO2_breakpoints = [(0, 35, 0, 50), (36, 75, 51, 100), (76, 185, 101, 150), 
                     (186, 304, 151, 200), (305, 604, 201, 300), (605, 9999, 301, 500)]
NO2_breakpoints = [(0, 53, 0, 50), (54, 100, 51, 100), (101, 360, 101, 150), 
                     (361, 649, 151, 200), (650, 1249, 201, 300), (1250, 9999, 301, 500)]
CO_breakpoints = [(0, 4, 0, 50), (5, 9, 51, 100), (10, 12, 101, 150), 
                     (13, 15, 151, 200), (16, 30, 201, 300), (31, 9999, 301, 500)]
O3_breakpoints = [(0, 50, 0, 50), (51, 70, 51, 100), (71, 85, 101, 150),
                    (86, 105, 151, 200), (106, 205, 201, 300), (206, 999999, 301, 500)]


```

```python
#fungsi untuk menentukan nilai aqi berdasarkan sub index
def calculate_sub_index(concentration, breakpoints):
    for (low_conc, high_conc, low_index, high_index) in breakpoints:
        if low_conc <= concentration <= high_conc:
            return low_index + (high_index - low_index) / (high_conc - low_conc) * (concentration - low_conc)
    # jika nilai sub index melebihi batas maka akan dikembalikan menjadi nilai tertinggi aqi
    return 500
```

```python
# menghitung nilai AQI untuk setiap baris pada data set
def calculate_aqi(row):
   
    # mengubah nilai konsentrasi polutan menjadi sub index
    row['PM2_5_sub_index'] = calculate_sub_index(row['PM2_5'], PM2_5_breakpoints)
    row['PM10_sub_index'] = calculate_sub_index(row['PM10'], PM10_breakpoints)
    row['NO2_sub_index'] = calculate_sub_index(row['NO2'], NO2_breakpoints)
    row['O3_sub_index'] = calculate_sub_index(row['O3'], O3_breakpoints)
    row['CO_sub_index'] = calculate_sub_index(row['CO'] * 0.001, CO_breakpoints)  # mengubah g/m menjadi mg/m
    row['SO2_sub_index'] = calculate_sub_index(row['SO2'], SO2_breakpoints)
 

    # AQI merupakan nilai sub index tertinggi
    aqi = max(row['PM2_5_sub_index'], row['PM10_sub_index'], row['NO2_sub_index'],
              row['O3_sub_index'], row['CO_sub_index'], row['SO2_sub_index'])
    return aqi
```

panduan membuat kode pemrogaman untuk menghitung aqi didapat dari [https://medium.com](/dare-to-be-better/case-study-4-a-step-by-step-guide-to-analyzing-the-aqi-air-quality-index-778c14b0dcfc) dan disesuaikan menggunakan pedoman penulisan dari EPA [https://www.airnow.gov](/sites/default/files/2020-05/aqi-technical-assistance-document-sept2018.pdf)

```python
#memasukkan kedalam dataframe aotizhongxin
aotizhongxin_df['AQI'] = aotizhongxin_df.apply(calculate_aqi, axis=1)
aotizhongxin_df.head()
```

```python
#memasukkan kedalam dataframe achangpin
changpin_df['AQI'] = changpin_df.apply(calculate_aqi, axis=1)
changpin_df.head()
```

```python
#memasukkan kedalam dataframe dongsi
dongsi_df['AQI'] = dongsi_df.apply(calculate_aqi, axis=1)
dongsi_df.head()
```

```python
#membuat kategori berdasarkan panduan aqi
aqi_categories = {
    (0, 50): 'Good',
    (51, 100): 'Moderate',
    (101, 150): 'Unhealthy for Sensitive Groups',
    (151, 200): 'Unhealthy',
    (201, 300): 'Very Unhealthy',
    (301, 500): 'Hazardous'
}
```

```python
#fungsi untuk mengelompokkan berdasarkan kategori aqi
def get_aqi_category(aqi):
    for (low_aqi, high_aqi), category in aqi_categories.items():
        if low_aqi <= aqi <= high_aqi:
            return category
    return 'Hazardous'
```

```python
#memasukkan kedalam data
aotizhongxin_df['AQI_Category'] = aotizhongxin_df['AQI'].apply(get_aqi_category)
aotizhongxin_df.head()
```

```python
#memasukkan kedalam data
changpin_df['AQI_Category'] = changpin_df['AQI'].apply(get_aqi_category)
changpin_df.head()
```

```python
#memasukkan kedalam data
dongsi_df['AQI_Category'] = dongsi_df['AQI'].apply(get_aqi_category)
dongsi_df.head()
```
